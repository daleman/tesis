```python
import json
tweets = []

for line in open('25.json'):
  try: 
    tweets.append(json.loads(line))
  except:
    pass
```

```python
print str(len(tweets[0].keys())) + " campos de tweet: \n" + str(tweets[0].keys()) + " en el tweet " + str(0) 
print "hay " + str(len(tweets)) + " tweets"
```

```python
ids = [tweet['id_str'] for tweet in tweets]
texts = [tweet['text'] for tweet in tweets]
times = [tweet['created_at'] for tweet in tweets]
print "\n" + "El id de cada tweet:\n" + str(ids)
print "\n" + "El texto de cada tweet:\n" + str(texts)
print "\n" + "El horario en que se creo de cada tweet:\n" + str(times)
print "\n" + "Los campos de cada user:\n" + str(tweet['user'].keys())
```

```python
screen_names = [tweet['user']['screen_name'] for tweet in tweets]
names = [tweet['user']['name'] for tweet in tweets]
print screen_names
print names
```

```python
lats = [(T['geo']['coordinates'][0] if T['geo'] else None) for T in tweets]
lons = [(T['geo']['coordinates'][1] if T['geo'] else None) for T in tweets]
print zip(lats,lons)
place_names = [(T['place']['full_name'] if T['place'] else None) for T in tweets]
place_types = [(T['place']['place_type'] if T['place'] else None) for T in tweets]
print place_names
```

## Armando un csv con los atrubutos identificados

```python
out = open('/tmp/tweets.csv', 'w')
print >> out, 'id,created,text,screen name,name,mention 1,mention 2,hashtag 1,hashtag 2,url 1,url 2,lat,lon,place name,place type'

rows = zip(ids, times, texts, screen_names, names, lats, lons, place_names, place_types)

from csv import writer
csv = writer(out)

for row in rows:
    values = [(value.encode('utf8') if hasattr(value, 'encode') else value) for value in row]
    csv.writerow(values)

out.close()

#mostramos en pantalla
for line in open('/tmp/tweets.csv'):
  try: 
    print line
  except:
    pass
```
